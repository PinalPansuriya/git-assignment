## 3)	Use pandas to create a dataframe that aggregates and sums both confirmed cases and deaths on a global level

```python
import pandas as pd

# URL for the daily confirmed cases and deaths data
confirmed_cases_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
deaths_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"

# Read the data into pandas dataframes
confirmed_cases_df = pd.read_csv(confirmed_cases_url)
deaths_df = pd.read_csv(deaths_url)

# Aggregate the sum of confirmed cases and deaths across all countries/regions
global_data = pd.DataFrame()
global_data["Confirmed"] = confirmed_cases_df.iloc[:, 4:].sum(axis=0)
global_data["Deaths"] = deaths_df.iloc[:, 4:].sum(axis=0)

# Display the aggregated data
global_data.head()
```

# Overall American Market(Standard & Poor's 500)


we use yahoo finance API for the fetching stock market data because alphavantage API not give proper data

```python
import yahoo_fin.stock_info as si
import plotly.graph_objects as go

# Define stock symbol for S&P/TSX Composite Index
index_symbol = "^GSPTSE"

# Get stock data from Yahoo Finance for a specific date range
stock_data = si.get_data(index_symbol, start_date="2018-12-01", end_date="2023-04-19")

# Define a function to calculate and display percentage change
def on_click(event):
    x = event['points'][0]['x']  # x-coordinate of clicked point
    y = event['points'][0]['y']  # y-coordinate of clicked point
    if x is not None and y is not None:
        # Calculate percentage change based on clicked y-coordinate
        prev_adjclose = stock_data.iloc[0]['adjclose']
        curr_adjclose = y
        percent_change = ((curr_adjclose - prev_adjclose) / prev_adjclose) * 100
        print(f"Percentage Change: {percent_change:.2f}%")

# Create a Plotly figure
fig = go.Figure()

# Add stock price trace to the figure
fig.add_trace(go.Scatter(x=stock_data.index, y=stock_data['adjclose'], mode='lines'))

# Update figure layout
fig.update_layout(
    title="American stock Market changes with COVID-19 Pandemic",
    xaxis_title="Date",
    yaxis_title="Adjusted Close Price"
)

# Connect click event to function
fig.data[0].on_click(on_click)

# Show the figure
fig.show()
```



*   This graph give information about the American stock market from Jan 2019 to till now. 
*  This graph also show the huge reflection of american market after the covid cases.


*  Overall , this graph indicates that when covid is come the stock market was decrease but after that it increses over the time.

# c)	Travel sector(Expedia Group Inc)


```python

import yahoo_fin.stock_info as si
import plotly.graph_objects as go

# Define stock symbols for travel sector companies
stock_symbols = ["EXPE"]

# Get stock data from Yahoo Finance for a specific date range for each stock symbol
stock_data = {}
for symbol in stock_symbols:
    stock_data[symbol] = si.get_data(symbol, start_date="2018-12-01", end_date="2023-04-19")

# Define a function to calculate and display percentage change
def on_click(event):
    x = event['points'][0]['x']  # x-coordinate of clicked point
    y = event['points'][0]['y']  # y-coordinate of clicked point
    if x is not None and y is not None:
        # Calculate percentage change based on clicked y-coordinate
        for symbol in stock_symbols:
            prev_adjclose = stock_data[symbol].iloc[0]['adjclose']
            curr_adjclose = y
            percent_change = ((curr_adjclose - prev_adjclose) / prev_adjclose) * 100
            print(f"Percentage Change for {symbol}: {percent_change:.2f}%")

# Create a Plotly figure
fig = go.Figure()

# Add stock price traces to the figure for each stock symbol
for symbol in stock_symbols:
    fig.add_trace(go.Scatter(x=stock_data[symbol].index, y=stock_data[symbol]['adjclose'], mode='lines', name=symbol))

# Update figure layout
fig.update_layout(
    title="Travel Sector Stock Prices change with COVID-19 Pandemic",
    xaxis_title="Date",
    yaxis_title="Adjusted Close Price"
)

# Connect click event to function
fig.data[0].on_click(on_click)

# Show the figure
fig.show()
```



*   This graph give information about the Travel sector stock market from Jan 2019 to till now.
*   Overall, this graph indicates that when covid came the stock market decreased after some time it starts to increase.However,from jan 2022 it started decrease.
